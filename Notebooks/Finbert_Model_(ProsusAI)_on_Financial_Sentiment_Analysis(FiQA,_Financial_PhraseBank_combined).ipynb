```python
import numpy as np
import pandas as pd
import transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
```

```python
dataset=pd.read_csv('data.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
finbert=AutoModelForSequenceClassification.from_pretrained('ProsusAI/finbert',num_labels=3)
tokenizer=AutoTokenizer.from_pretrained('ProsusAI/finbert')
nlp=pipeline('sentiment-analysis',model=finbert,tokenizer=tokenizer)
results=dataset['Sentence'].apply(lambda x: nlp(x)[0])
dataset['pred_Sentiment']=results.apply(lambda x: x['label'].lower())
```

```python
dataset
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(dataset['Sentiment'],dataset['pred_Sentiment'])
accuracy=accuracy_score(dataset['Sentiment'],dataset['pred_Sentiment'])
print(cm)
print(accuracy)
```
